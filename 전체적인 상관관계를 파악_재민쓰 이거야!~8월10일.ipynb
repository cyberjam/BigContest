```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 모든 파일 불러오기
activity = pd.read_csv('C:/Users/CPB06GameN/PycharmProjects/github/bigcon/train_activity.csv')
combat = pd.read_csv('C:/Users/CPB06GameN/PycharmProjects/github/bigcon/train_combat.csv')
label = pd.read_csv('C:/Users/CPB06GameN/PycharmProjects/github/bigcon/train_label.csv')
payment = pd.read_csv('C:/Users/CPB06GameN/PycharmProjects/github/bigcon/train_payment.csv')
pledge = pd.read_csv('C:/Users/CPB06GameN/PycharmProjects/github/bigcon/train_pledge.csv')
trade = pd.read_csv('C:/Users/CPB06GameN/PycharmProjects/github/bigcon/train_trade.csv')
```

```python
# 아이템을 주고 받는 아이디가 동일하지 않기때문에 주는 유저로 기준을 잡았다.

source_acc_id=trade.loc[:,['source_acc_id','type','item_amount','item_price']]
# source_acc_id.tail()
```

```python
# acc_id 로 merge할 준비

# rename 하기(동일한 컬럼명)
source_acc_id=source_acc_id.rename(columns={'source_acc_id' : 'acc_id'})

# sum으로 묶었다(label 제외)
activity.groupby('acc_id').sum()
combat.groupby('acc_id').sum()
payment.groupby('acc_id').sum()
source_acc_id.groupby('acc_id').sum()
# source_acc_id.tail()

# 메모리 에러가 떠서 임의로 3000개를 잘랐다.
activity=activity[:3001]
combat=combat[:3001]
label=label[:3001]
payment=payment[:3001]
pledge=pledge[:3001]
source_acc_id=source_acc_id[:3001]
```

```python
# 유저 아이디가 동일하게 들어있는 것이 아니기 때문에 우선적으로 outer로 묶어 놓고 
# 다음에 다시 acc_id로 groupby를 했다.
total_df=pd.merge(activity,combat, how='outer')
total_df=pd.merge(total_df,label, how='outer')
total_df=pd.merge(total_df,payment, how='outer')
total_df=pd.merge(total_df,pledge, how='outer')
total_df=pd.merge(total_df,source_acc_id, how='outer')
```

```python
#  acc_id로 묶어서 상관관계를 파악
# 여기서는 mean으로 하나 sum으로 하나 상관없다고 판단
a=total_df.groupby('acc_id').sum()
a.shape    # (8660, 35)
result=a.corr()
```
