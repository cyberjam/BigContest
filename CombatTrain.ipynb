```python
# from pandas.io.parsers import read_csv
import pandas as pd

Big_df = pd.read_csv("../Data2e/BigContest/train_combat.csv")
# print("Dataframe",df)
Big_df
```

```python
# https://cleancode-ws.tistory.com/63 pandas 결측치 찾기

Big_df_na_sum = Big_df.isnull().sum()
print(Big_df_na_sum)
```

```python
#전투활동을 한번이라도 한 사람

#https://m.blog.naver.com/PostView.nhn?blogId=ossiriand&logNo=220703077723&proxyReferer=https%3A%2F%2Fwww.google.com%2F
# 해당 조건 충족하는 행 추출하기

combat=Big_df[(Big_df['pledge_cnt']>0)|(Big_df['random_attacker_cnt']>0)|(Big_df['random_defender_cnt']>0)|(Big_df['temp_cnt']>0)|(Big_df['same_pledge_cnt']>0)|(Big_df['etc_cnt']>0)|(Big_df['num_opponent']>0)]
# 437036 rows × 13 columns
combat
```

```python
combat[combat['acc_id']==62835]
```
