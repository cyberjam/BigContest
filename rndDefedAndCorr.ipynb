## Day2 가설 : 막피 당한 유저는 이탈률이 높을까?

```python
import pandas as pd
```

```python
combat = pd.read_csv('BigContest/train_combat.csv')
```

```python
combat.dtypes
```

```python
# 날짜, 유저, 캐릭터, 막피 공격 받은 횟수
rnd_defen= combat[['day','acc_id','char_id','random_defender_cnt']]
rnd_defen
```

```python
# 막피 공격 받은 사람만
rnd_defen = rnd_defen[rnd_defen['random_defender_cnt']>0]
rnd_defen
```

```python
# 아이디, 날짜 순서로 내림차순 정렬
rnd_defen = rnd_defen.sort_values(['acc_id','day'],ascending=True)
rnd_defen
```

```python
# 막피를 가장 많이 당한 사람이 누굴까
rnd_defen[ rnd_defen['random_defender_cnt'] ==rnd_defen['random_defender_cnt'].max()]
```

```python
# 후보1 
# 하루 타격이 가장 컸던 유저 95번 맞음   112.391126 /1.183064
rnd_defen[ rnd_defen['acc_id'] ==83876]
```

```python
# 후보2 
# 평균적으로 많이 쳐맞음
# 평균은 쳐맞은 일별 평균

rnd_defen[ rnd_defen['acc_id'] ==123532]
```

346526 유저를 봤을 때 누적이 아니다.

```python
# 쳐맞은 횟수와 타격을 고려한 평균 # 쳐맞은 일별 평균
grp_mean_rnd_defen = rnd_defen.groupby(['acc_id','char_id']).mean()
grp_mean_rnd_defen
```

```python
grp_mean_rnd_defen_dro_day = grp_mean_rnd_defen.drop(columns = 'day')
grp_mean_rnd_defen_dro_day.sort_values('random_defender_cnt',ascending=False)
```

```python
# 모든 유저에게 같은 일수로 나누자
```

```python
# 많이 쳐맞은 사람이 이탈했을까
label = pd.read_csv('BigContest/train_label.csv')
```

```python
# 총량으로 하자
grp_sum_rnd_defen = rnd_defen.groupby(['acc_id','char_id']).sum().sort_values('random_defender_cnt',ascending=False).drop(columns='day')
grp_sum_rnd_defen
```

```python
# 새로운 인물 등장
```

```python
label[label['acc_id']==52326]
```

```python
#오 대박 생존
```

```python
sum_mrg= pd.merge(grp_sum_rnd_defen,label, how = 'inner', on = 'acc_id').sort_values('random_defender_cnt',ascending=False).drop(columns='amount_spent')

sum_mrg
```

```python

rdm_defcnt=sum_mrg[['random_defender_cnt']]
rdm_defcnt
```

```python

srv_tm=sum_mrg[['survival_time']]
srv_tm
```

```python
srv_tm['survival_time']
```

```python
grp_df = pd.DataFrame({'survival_time':srv_tm['survival_time'], 'random_defender_cnt':rdm_defcnt['random_defender_cnt']})
```

```python
grp_df.plot.scatter(x='survival_time',y='random_defender_cnt')
```

```python
# 10일 살았던 사람이나, 15일 살았던 사람이나, 30일 살았던 사람이나, 40일 살았던 사람이나 맞은 횟수는 비슷하다.
#오히려 64일 이상 생존했던 사람들이 많이 맞았다
#그만큼 많이 때렸겠지?
# 서로 때리며 다독이는 걸까
```
