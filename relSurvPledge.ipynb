```python


# 단 혈맹을 하지 않는 유저들은 우선 제외.
# 생존데이터와 혈맹 데이터 병합

df_pledge_label=pd.merge(df_pledge,df_label,how='inner',on='acc_id') 
# pd.merge(df_pledge,df_label, how = 'outer')
# pd.merge(df_pledge,df_label, how = 'outer')
# df_pledge_label[df_pledge_label.acc_id==78896]
df_pledge_label
```

```python
# 질문? : 밑에 것으로 한다면 혈맹을 제외한 모든 것이 평균이 나와 버린다.
# df_pledge_label.groupby(['pledge_id']).mean()
```

```python
df_drop=df_pledge_label.groupby(['pledge_id'])[['survival_time']].mean() # 혈맹내 캐릭터 수로 생존률을 구함.
df_drop
# 각 혈맹의 평균 생존율
```

```python
combat = pd.read_csv('BigContest/train_combat.csv')
label = pd.read_csv('BigContest/train_label.csv')
payment = pd.read_csv('BigContest/train_payment.csv')
pledge = pd.read_csv('BigContest/train_pledge.csv')
trade = pd.read_csv('BigContest/train_trade.csv')
file_list = [combat ,
label ,
payment ,
pledge,
trade ]

```

```python
combat_mean = combat.groupby('char_id').mean()
```

```python
pledge_mean = pledge.groupby(['pledge_id','char_id'],as_index=False).mean()
pledge_mean
```

```python
result_pledge= df_drop.merge(pledge_mean,on='pledge_id')
result_pledge
```

```python
result_combat_pledge= result_pledge.merge(combat,on='char_id')
result_combat_pledge
```

```python
result=result_combat_pledge.groupby(['pledge_id','survival_time'],as_index=False).mean()
result
```
