## Load Data

```python
activity_raw = pd.read_csv('../BigContest/train_activity.csv')
label_raw = pd.read_csv('../BigContest/train_label.csv')
```

```python
activity_raw.dtypes
```

```python
activity = activity_raw
label = label_raw
```

```python
activity_grp = activity.groupby('acc_id',as_index=False).sum()
```

```python
activity_grp.shape
```

```python
mrg_act_lbl = activity_grp.merge(label,on = 'acc_id')
```

```python
mrg_act_lbl.shape
```

```python
mrg_act_lbl.head()
```

```python
#훈련세트, 테스트세트 나누기
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(mrg_act_lbl, test_size = 0.3)

train_set.shape
test_set.shape
```

```python
mrg_act_lbl_train = train_set
mrg_act_lbl_train.head()
```

```python
mrg_act_lbl_test = test_set
mrg_act_lbl_test.head()
```

## Data Preprocessing

```python
predictors = ["playtime", "fishing", "private_shop"]
x = mrg_act_lbl[predictors]
y = mrg_act_lbl["survival_time"]
```

## DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision = DecisionTreeClassifier().fit(x,y)
print(decision)
```

## RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forst = RandomForestClassifier(n_jobs=-1).fit(x,y)
forst
```

## Cross Validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
x
```

```python
y
```

```python
for i in range(50):
    decision_score = cross_val_score(decision, x, y, cv=5).mean()
    forst_score = cross_val_score(forst, x, y, cv=5).mean()
    print('='*10)
    print("DecisionTree = {0:.6f}".format(decision_score))
    print("RandomForest = {0:.6f}".format(forst_score))
    print('='*10)
```
