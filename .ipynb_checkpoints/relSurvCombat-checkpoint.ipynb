# 2) Standardization

```python
grp_df["survival_time"].mean(), grp_df["survival_time"].std()
```

```python
grp_df["survival_time"] = ( grp_df["survival_time"] - grp_df["survival_time"].mean() ) / (grp_df["survival_time"].std() )
```

```python
grp_df
```

```python
grp_df["amount_spent"].mean(), grp_df["amount_spent"].std()
grp_df["amount_spent"] = ( grp_df["amount_spent"] - grp_df["amount_spent"].mean() ) / (grp_df["survival_time"].std() )
grp_df
```
