# 파생변수 생성 ( 막피 방어+ 막피 공격)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
combat_raw = pd.read_csv('BigContest/train_combat.csv')
combat=combat_raw
```

```python
combat.dtypes
```

```python
att = combat[['random_attacker_cnt','random_defender_cnt','acc_id']]
```

```python
att
```

```python
combat['random_attacker_cnt'][12]+ combat['random_defender_cnt'][12]
```

```python
att.summ = att.random_attacker_cnt + att.random_defender_cnt
```

```python
att.dtypes
```

```python
att
```

```python
label_raw = pd.read_csv('BigContest/train_label.csv')
label = label_raw
label
```

```python
result = label[['acc_id','survival_time']].merge(att,how='inner',on = 'acc_id').sort_values('sum',ascending=False)
result
```

```python
result.dtypes
```

```python


plt.hist(result.survival_time,10, density=True, facecolor='g', alpha=0.75,rwidth = 0.8)

plt.xlabel('survival_time')
plt.ylabel('sum')
plt.title('Random_Sum')
# plt.text(64, 0.25, r'$\mu=100,\ \sigma=15$')
plt.axis([1, 63, 0, 0.1])
plt.grid(True)
```

```python
grp_df = pd.DataFrame({'survival_time':result.survival_time, 'cnt':result['sum']})
grp_df.plot.scatter(x='survival_time',y='cnt')
plt.show()
```
